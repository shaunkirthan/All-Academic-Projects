# RFM Calculation

```python
import datetime as dt
recency=((dt.datetime(2011,12,10).date()-data.groupby("CustomerID")["InvoiceDate"].max().dt.date).dt.days).reset_index()
recency.rename(columns={'InvoiceDate': 'Recency'}, inplace=True)
data=data.merge(recency,on="CustomerID",how="left")
data
```

```python
frequency=(data.groupby("CustomerID")["InvoiceNo"].count()).reset_index()
frequency.rename(columns={'InvoiceNo': 'Frequency'}, inplace=True)
data=data.merge(frequency,on="CustomerID",how="left")
data
```

```python
data["Total order value"]=data["UnitPrice"]*data["Quantity"]
monetary=(data.groupby("CustomerID")["Total order value"].sum()).reset_index()
monetary.rename(columns={'Total order value': 'Monetary'}, inplace=True)
data=data.merge(monetary,on="CustomerID",how="left")
data
```

# RFM Segmentation

```python
recency_scores = [5,4, 3, 2, 1]  # Higher score for lower recency (more recent)
frequency_scores = [1, 2, 3, 4,5]  # Higher score for higher frequency
monetary_scores = [1, 2, 3, 4,5]  # Higher score for higher monetary value

# Calculating RFM scores
data['RecencyScore'] = pd.cut(data['Recency'], bins=5, labels=recency_scores)
data['FrequencyScore'] = pd.cut(data['Frequency'], bins=5, labels=frequency_scores)
data['MonetaryScore'] = pd.cut(data['Monetary'], bins=5, labels=monetary_scores)
```

```python
data['RecencyScore'] = data['RecencyScore'].astype(int)
data['FrequencyScore'] = data['FrequencyScore'].astype(int)
data['MonetaryScore'] = data['MonetaryScore'].astype(int)
data['RFM_Score'] = data['RecencyScore'] + data['FrequencyScore'] + data['MonetaryScore']
data
```

# Customer Segmentation

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(data[["Recency","Frequency","Monetary"]])

sum_of_sq_dist = {}
for k in range(1,11):
    km = KMeans(n_clusters= k, init= 'k-means++', max_iter= 300, n_init=10)
    km = km.fit(scaled)
    sum_of_sq_dist[k] = km.inertia_
    
#Plot the graph for the sum of square distance values and Number of Clusters
sns.pointplot(x = list(sum_of_sq_dist.keys()), y = list(sum_of_sq_dist.values()))
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Sum of Square Distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# we chose n_clusters as 3 because that is our elbow point
KMean_clust = KMeans(n_clusters= 3, init= 'k-means++', max_iter= 300, n_init=10)
KMean_clust.fit(scaled)

#Find the clusters for the observation given in the dataset
data['Cluster'] = KMean_clust.labels_
plt.figure(figsize=(7, 7))

# Scatter Plot Frequency Vs Recency
Colors = ["violet", "grey", "pink"]
data['Color'] = data['Cluster'].map(lambda p: Colors[p])

for cluster, color in zip(range(len(Colors)), Colors):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(
        x=cluster_data["Recency"],
        y=cluster_data["Frequency"],
        c=color,
        label=f'Cluster {cluster}'
    )

# Add labels and title
plt.xlabel("Recency")
plt.ylabel("Frequency")
plt.title("Scatter Plot: Frequency vs Recency")

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Define custom stop words (articles and connectors)
custom_stop_words = ['this', 'is', 'and', 'or', 'but', 'and', 'the', 'a', 'an']  # Add more as needed

# Step 1: Create a Bag of Words representation with a custom token pattern and stop words
vectorizer = CountVectorizer(token_pattern=r'\b[^\d\W]+\b', stop_words=custom_stop_words)
bow_matrix = vectorizer.fit_transform(categories100['Description'])

# Convert BoW matrix to DataFrame
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Step 2: Assign the most occurred word in each row to a new column
most_occurred_word = bow_df.idxmax(axis=1)
most_occurred_word[bow_df.sum(axis=1) == 0] = ''  # Handle rows with all zero values
categories100['Category'] = most_occurred_word

print(categories100)
```

```python
highest_revenue_category=categories100.groupby("Category")["Total order value"].max().reset_index()
highest_revenue_category=highest_revenue_category.sort_values(by='Total order value', ascending=False)
print("Product category with the highest revenue generation is:")
highest_revenue_category.head(1)
```
